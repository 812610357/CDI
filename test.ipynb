```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,1536)
y1=np.linspace(0.5,1)
y2=np.linspace(1,0.5)

plt.axis([0, 1536, 0, 1])

plt.plot(x,y1,label="High-pass")
plt.plot(x,y2,label="Low-pass")

plt.xlabel("Frequence")  # X
plt.legend()  # 
plt.title("Fliter")  # 
```
